```python
import pandas as pd
```

```python
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
#loading data
dataset = load_dataset ('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.iloc[0]
```

```python
# .loc with subsetting nan values (cleaning data)
df.loc[:, "salary_rate": "salary_hour_avg"].dropna(subset='salary_rate')
```

```python
df.pivot_table(values='salary_hour_avg', index='job_country', columns='job_title_short', aggfunc='median')
```

```python
df.pivot_table(values='job_title_short', index='job_country', columns='job_title_short', aggfunc='count')
```

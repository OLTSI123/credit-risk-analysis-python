```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90, autopct='%1.1f%%')
plt.title('Work from home Status')
plt.ylabel("")
plt.show()
```

```python
fig, ax = plt.subplots(1, 3)

dict_column = {
    'job_work_from_home': 'Work from Home',
    'job_no_degree_mention': 'Job Degree Req.',
    'job_health_insurance': 'health_insurance_offered'
}

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle=90, autopct='%1.1f%%',  labels=['False', 'True'])
    ax[i].set_title(title)
plt.show()
```

```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df.info()
```

```python
df['job_health_insurance'].value_counts().plot(kind='pie', startangle=90, autopct='%1.1f%%')
plt.title('Health insurance')
plt.ylabel("")
plt.show()
```

```python
df['job_schedule_type'].value_counts().head().plot(kind='barh')
plt.title('job_schedule_type')
plt.ylabel("")
plt.show()
```

```python
df['job_no_degree_mention'].value_counts().plot(kind='pie', startangle=90, autopct='%1.1f%%')
plt.title('Health insurance')
plt.ylabel("")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2)

dict_column = {
    'job_no_degree_mention': 'Job Degree req',
    'job_health_insurance': 'Health insurance',
}

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle=90, autopct='%1.1f%%',  labels=['False', 'True'])
    ax[i].set_title(title)
plt.show()
```

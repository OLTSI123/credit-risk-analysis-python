```python
import pandas as pd
import re

# ============================
# 1️⃣ Read CSV properly
# ============================

# Adjust sep=',' or sep=';' depending on your file
df_raw = pd.read_csv("Transactions2.csv", sep=',', encoding='cp1252', dtype=str)

# ============================
# 2️⃣ Inspect raw headers
# ============================
print("Raw headers:")
for col in df_raw.columns:
    print(repr(col))

# ============================
# 3️⃣ Clean headers
# ============================
df_raw.columns = [col.strip().replace('\xa0','') for col in df_raw.columns]

# Remove any empty columns
df_raw = df_raw.loc[:, df_raw.columns.str.strip() != '']

print("\nCleaned headers:")
print(df_raw.columns.tolist())

# ============================
# 4️⃣ Define European number parser
# ============================
def parse_euro_number(s):
    if pd.isna(s):
        return 0.0
    s = str(s)
    s = s.replace('−', '-')       # Unicode minus → ASCII
    s = s.replace(' ', '').replace('\xa0','')  # remove spaces/non-breaking
    s = re.sub(r'\.(?=\d{3},)', '', s)        # remove thousands dots
    s = s.replace(',', '.')                     # decimal comma → dot
    if s.startswith('(') and s.endswith(')'):   # parentheses as negative
        s = '-' + s[1:-1]
    try:
        return float(s)
    except:
        return 0.0

# ============================
# 5️⃣ Convert numeric columns
# ============================
numeric_cols = [
    "Value EUR",
    "AutoFX Fee",
    "Transaction and/or third party fees EUR",
    "Kokonaissumma EUR",
    "Quantity",
    "Kurssi",
    "Markkina-arvo",
    "Vaihtokurssi"
]

for col in numeric_cols:
    if col in df_raw.columns:
        df_raw[col] = df_raw[col].apply(parse_euro_number)
    else:
        print(f"Warning: Column '{col}' not found in CSV")

# ============================
# 6️⃣ Recalculate totals
# ============================
if "Value EUR" in df_raw.columns and "Transaction and/or third party fees EUR" in df_raw.columns:
    df_raw["Kokonaissumma EUR"] = df_raw["Value EUR"] + df_raw["Transaction and/or third party fees EUR"]

# ============================
# 7️⃣ Check results
# ============================
print("\nSample numeric data:")
print(df_raw[numeric_cols].head(20))

# ============================
# 8️⃣ Save cleaned CSV (optional)
# ============================
df_raw.to_csv("Transactions2_cleaned.csv", index=False)
```

```python
import re
import pandas as pd

def parse_euro_number(s):
    """
    Convert a European-formatted number to float:
    - Unicode minus → ASCII minus
    - Decimal comma → dot
    - Remove thousands separators
    - Handle parentheses as negative numbers
    """
    if pd.isna(s):
        return 0.0
    s = str(s)
    s = s.replace('−', '-')        # Unicode minus → ASCII
    s = s.replace(' ', '').replace('\xa0','')  # remove spaces/non-breaking
    s = re.sub(r'\.(?=\d{3},)', '', s)        # remove thousands separator dots
    s = s.replace(',', '.')                      # decimal comma → dot
    if s.startswith('(') and s.endswith(')'):    # parentheses as negative
        s = '-' + s[1:-1]
    try:
        return float(s)
    except:
        return 0.0
```

```python
numeric_cols = [
    "Value EUR",
    "AutoFX Fee",
    "Transaction and/or third party fees EUR",
    "Kokonaissumma EUR",
    "Markkina-arvo",
    "Quantity",
    "Kurssi"
]

# Add any Unnamed numeric columns as well
numeric_cols += [col for col in df_raw.columns if "Unnamed" in col]

# Apply parser
for col in numeric_cols:
    if col in df_raw.columns:
        df_raw[col] = df_raw[col].apply(parse_euro_number)
```

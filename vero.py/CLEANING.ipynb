```python
cols_to_fix = ["Value EUR", "AutoFX Fee", "Transaction and/or third party fees EUR", "Kokonaissumma EUR"]

def fix_european_number(s):
    if pd.isna(s):
        return s
    s = str(s)
    s = s.replace('\u2212', '-')  # Unicode minus → ASCII minus
    s = s.replace('\u2013', '-').replace('\u2014', '-')  # just in case
    s = s.replace('\xa0', '').replace(' ', '')  # remove spaces
    # Remove thousands separator dots (before changing decimal comma)
    s = s.replace('.', '')  
    # Replace decimal comma with dot
    s = s.replace(',', '.')
    return s

for col in cols_to_fix:
    df[col] = df[col].apply(fix_european_number)
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Step 1: Try reading with UTF-16 (common for Excel exports)
df_raw = pd.read_csv("Transactions2.csv", dtype=str, encoding='utf-16', sep=';')  # or 'utf-16le', try both

# Step 2: Inspect raw problematic cells
print(repr(df_raw.loc[0, 'Value EUR']))  # shows all invisible characters

# Step 3: Universal clean function
import re

def robust_parse_euro_number(s):
    if pd.isna(s):
        return 0.0
    s = str(s)
    # Keep only digits, minus, comma, dot
    s = re.sub(r'[^0-9,\-\.()]', '', s)
    # Convert parentheses to minus
    if s.startswith('(') and s.endswith(')'):
        s = '-' + s[1:-1]
    # Replace Unicode minus with ASCII minus
    s = s.replace('−', '-')
    # Remove thousands separator dots
    s = re.sub(r'\.(?=\d{3},)', '', s)
    # Replace decimal comma with dot
    s = s.replace(',', '.')
    # Convert to float
    try:
        return float(s)
    except:
        return 0.0

# Step 4: Apply to all relevant columns
cols_to_fix = ["Value EUR", "AutoFX Fee", "Transaction and/or third party fees EUR", "Kokonaissumma EUR"]

for col in cols_to_fix:
    df_raw[col] = df_raw[col].apply(robust_parse_euro_number)

df_raw["Kokonaissumma EUR"] = df_raw["Value EUR"] + df_raw["Transaction and/or third party fees EUR"]

# Check results
print(df_raw[cols_to_fix].head(20))
```

```python
import pandas as pd
from io import StringIO

file_path = r"C:\Users\aarni\Downloads\Transactions.csv"

# Read raw file with utf-8-sig to automatically remove BOM
with open(file_path, "r", encoding="utf-8-sig") as f:
    text = f.read()

# Fix broken minus signs
text = text.replace("âˆ’", "-")

# Remove outer quotes around rows
lines = text.splitlines()
cleaned_lines = []

for line in lines:
    line = line.strip()
    if line.startswith('"') and line.endswith('"'):
        line = line[1:-1]
    cleaned_lines.append(line)

text = "\n".join(cleaned_lines)

# Replace double double-quotes
text = text.replace('""', '')

# Load into pandas
df = pd.read_csv(StringIO(text), sep=",", engine="python")

# Strip column names and values
df.columns = df.columns.str.strip()
for col in df.select_dtypes("object").columns:
    df[col] = df[col].str.strip()

# Remove unnecessary Unnamed columns
df = df.loc[:, ~df.columns.str.contains("Unnamed")]

print(df.head())
print(df.dtypes)
```

```python
import pandas as pd

file_path = r"C:\Users\aarni\Downloads\Transactions.csv"

# Read the raw file
with open(file_path, "r", encoding="utf-8-sig") as f:
    lines = f.readlines()

rows = []

for line in lines:
    line = line.strip()
    if line.startswith('"') and line.endswith('"'):
        line = line[1:-1]  # remove outer quotes
    line = line.replace('""', '')  # remove double quotes inside
    # Split by comma
    parts = line.split(',')
    rows.append(parts)

# Use first row as header
header = rows[0]
data = rows[1:]

# Create dataframe
df = pd.DataFrame(data, columns=header)

# Strip whitespace
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

print(df.head())
```
